```python
def gender_feature(word):
    return{'last_letter' :word[-2:]}
```

```python
gender_feature("Ankush")
```

```python
from nltk.corpus import names
labeled_names = ([(name,'male') for name in names.words('male.txt')]+[(name,'female') for name in names.words('female.txt')])
```

```python
labeled_names
```

```python
len(labeled_names)
```

```python
import random
random.shuffle(labeled_names)
```

```python
labeled_names
```

```python
feature_set=[(gender_feature(n),gender) for (n,gender) in labeled_names]
```

```python
feature_set
```

```python
train_set,test_set=feature_set=feature_set[:5000],feature_set[5000:]
```

```python
import nltk
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
classifier.classify(gender_feature('David'))
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```
