```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = "https://www.maayboli.com/node/68966"
web_page = urlopen(url)
s=[]
soup = BeautifulSoup(web_page, 'html.parser')

for extract_div in soup.findAll("div", {"class" : "field-items"}):
        s.append(extract_div.text.split())
print(s)
```
